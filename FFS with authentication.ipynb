```python
from tkinter import *
from functools import partial

def validateLogin(username, password):
	print("username entered :", username.get())
	print("password entered :", password.get())
	return

#window
tkWindow = Tk()  
tkWindow.geometry('400x350')  
tkWindow.title('User Login')

#username label and text entry box
usernameLabel = Label(tkWindow, text="User Name").grid(row=0, column=0)
username = StringVar()
usernameEntry = Entry(tkWindow, textvariable=username).grid(row=0, column=1)  

#password label and password entry box
passwordLabel = Label(tkWindow,text="Password").grid(row=1, column=0)  
password = StringVar()
passwordEntry = Entry(tkWindow, textvariable=password, show='*').grid(row=1, column=1)  

validateLogin = partial(validateLogin, username, password)

#login button
loginButton = Button(tkWindow, text="Login", command=validateLogin).grid(row=4, column=0)  

tkWindow.mainloop()



from tkinter import *
from tkinter import ttk
from tkinter import messagebox



root = Tk()
frame_header = ttk.Frame(root)
frame_header.pack()
headerlabel = ttk.Label(frame_header, text='FACULTY FEEDBACK FORM', foreground='BLACK',
                        font=('Arial', 24))
headerlabel.grid(row=0, column=1)
messagelabel = ttk.Label(frame_header,
                         text='PLEASE TELL US WHAT YOU THINK',
                         foreground='green', font=('Arial', 10))
messagelabel.grid(row=1, column=1)

frame_content = ttk.Frame(root)
frame_content.pack()
# def submit():
#     username = entry_name.get()
#     print(username)
myvar = StringVar()
var = StringVar()
var1 = StringVar()
var2 = StringVar()
var3 = StringVar()
var4 = StringVar()
var5 = StringVar()
var6 = StringVar()
# cmnt= StringVar()
namelabel = ttk.Label(frame_content, text='Name')
namelabel.grid(row=0, column=0, sticky='sw')
entry_name = ttk.Entry(frame_content, width=18, font=('Arial', 10), textvariable=myvar)
entry_name.grid(row=1, column=0, sticky='sw')

facultynamelabel = ttk.Label(frame_content, text='Faculty Name')
facultynamelabel.grid(row=0, column=1, sticky='sw')
entry_facultyname = ttk.Entry(frame_content, width=18, font=('Arial', 10), textvariable=var)
entry_facultyname.grid(row=1, column=1, sticky='sw')

commentlabel = ttk.Label(frame_content, text='Comment', font=('Arial', 14))
commentlabel.grid(row=12, column=0, sticky='sw')
textcomment = Text(frame_content, width=30, height=1)
textcomment.grid(row=12, column=0, columnspan=2)

branchlabel = ttk.Label(frame_content, text='Branch')
branchlabel.grid(row=2, column=0, sticky='sw')
entry_branch = ttk.Entry(frame_content, width=18, font=('Arial', 10), textvariable=var1)
entry_branch.grid(row=3, column=0, sticky='sw')

subjectlabel = ttk.Label(frame_content, text='Subject')
subjectlabel.grid(row=2, column=1, sticky='sw')
entry_subject = ttk.Entry(frame_content, width=18, font=('Arial', 10), textvariable=var2)
entry_subject.grid(row=3, column=1)

giveyourfeedbackonscalelabel = ttk.Label(frame_content, text='GIVE YOUR FEEDBACK ON SCALE', foreground='green' , font=('Arial', 16))
giveyourfeedbackonscalelabel.grid(row=6, column=0, sticky='sw')
entry_giveyourfeedbackonscale = ttk.Entry(frame_content, width=18, textvariable=var)

wayofteachinglabel = ttk.Label(frame_content, text='Way of teaching')
wayofteachinglabel.grid(row=7, column=0, sticky='sw')
entry_wayofteaching = ttk.Entry(frame_content, width=5, font=('Arial', 10), textvariable=var3)
entry_wayofteaching.grid(row=7, column=0)

interactionlabel = ttk.Label(frame_content, text='INTERACTION')
interactionlabel.grid(row=8, column=0, sticky='sw')
entry_interaction = ttk.Entry(frame_content, width=5, font=('Arial', 10), textvariable=var4)
entry_interaction.grid(row=8, column=0)

punctualitylabel = ttk.Label(frame_content, text='PUNCTUALITY')
punctualitylabel.grid(row=9, column=0, sticky='sw')
entry_punctuality = ttk.Entry(frame_content, width=5, font=('Arial', 10), textvariable=var5)
entry_punctuality.grid(row=9, column=0)

completionofsyllabuslabel = ttk.Label(frame_content, text='COMPLETION OF SYLLABUS')
completionofsyllabuslabel.grid(row=10, column=0, sticky='sw')
entry_completionofsyllabus = ttk.Entry(frame_content, width=5, font=('Arial', 10), textvariable=var6)
entry_completionofsyllabus.grid(row=10, column=0)


textcomment.config(wrap ='word')
# def clear():
#     textcomment.delete(1.0,'end')
def clear():
    global entry_name
    global entry_facultyname
    global textcomment
    global entry_branch
    global entry_subject
    global entry_wayofteaching
    global entry_interaction
    global entry_punctuality
    global entry_completionofsyllabus
    messagebox.showinfo(title='clear', message='Do you want to clear?')
    entry_name.delete(0, END)
    entry_facultyname.delete(0, END)
    textcomment.delete(1.0, END)
    entry_branch.delete(0, END)
    entry_subject.delete(0, END)
    entry_wayofteaching.delete(0, END)
    entry_interaction.delete(0, END)
    entry_punctuality.delete(0, END)
    entry_completionofsyllabus.delete(0, END)

def submit():
    global entry_name
    global entry_facultyname
    global textcomment
    global entry_branch
    global entry_subject
    global entry_wayofteaching
    global entry_interaction
    global entry_punctuality
    global entry_completionofsyllabus
    print('Name:{}'.format(myvar.get()))
    print('Faculty Name:{}'.format(var.get()))
    print('Branch:{}'.format(var1.get()))
    print('Subject:{}'.format(var2.get()))  
    print('Way of teaching:{}'.format(var3.get()))
    print('Interaction:{}'.format(var4.get()))
    print('Punctuality:{}'.format(var5.get()))
    print('Completion of syllabus:{}'.format(var6.get()))
    print('Comment:{}'.format(textcomment.get(1.0, END)))
    messagebox.showinfo(title='Submit', message='Thank you for your Feedback, Your Comments are Submited')
    entry_name.delete(0, END)
    entry_facultyname.delete(0, END)
    textcomment.delete(1.0, END)
    entry_branch.delete(0, END)
    entry_subject.delete(0, END)
    entry_wayofteaching.delete(0, END)
    entry_interaction.delete(0, END)
    entry_punctuality.delete(0, END)
    entry_completionofsyllabus.delete(0, END)


submitbutton = ttk.Button(frame_content, text='Submit', command=submit).grid(row=15, column=0, sticky='e')
clearbutton = ttk.Button(frame_content, text='Clear', command=clear).grid(row=15, column=1, sticky='w')

mainloop()
```
